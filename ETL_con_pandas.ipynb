```python
import pandas as pd
df=pd.read_csv("/content/data.csv", delimiter=',', encoding='unicode_escape')
print(df)
```

```python
df.info()
```

```python
df = df.rename(columns={'Country':'Country_Code', 'CustomerID': 'Customer_ID'})
```

```python
df.head()
```

```python
df['Customer_ID'] = df['Customer_ID'].astype(str)
```

```python
df.head()
```

```python
df.dropna(subset={'Customer_ID'}, inplace=True)
```

```python
def agg_zero(x):
    if len(x) <= 8:
      return '0' + x
    else:
      return x
```

```python
df['Customer_ID'] = df['Customer_ID'].apply(agg_zero)
```

```python
df.head()
```

```python
df['Customer_ID'] = df['Customer_ID'].apply(lambda x: x.rstrip('.0'))
```

```python
df.head()
```

```python
df['Segementacion']=df['StockCode'].apply(lambda x: 'VIP' if x.endswith('A') else 'None')
```

```python
vip_count = df['Segementacion'].value_counts()['VIP']
print(vip_count)
```
